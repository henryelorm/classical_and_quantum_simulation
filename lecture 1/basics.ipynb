# CACULATORS

```python
from ase.calculators.eam import EAM
from ase.calculators.emt import EMT
from gpaw import GPAW

from ase.build import bulk
from ase.visualize import view
from ase.io.vasp import read_vasp
```

```python
cu = bulk(name='Cu', crystalstructure='fcc', a=3.58, cubic=True, )

view(cu)
```

##### using in-built calculators eg. EMT

### Optimizers

### BFGS

```python
import numpy as np


from ase.optimize import BFGS
```

```python
d = 1.0
t = np.pi / 180 * 104.51
water = Atoms('H2O',
              positions=[(d, 0, 0),
                         (d * np.cos(t), d * np.sin(t), 0),
                         (0, 0, 0)])


# view(water)
```

```python
water.calc =  EMT()

water.get_potential_energy()
```

```python
dyn = BFGS(atoms=water)

dyn.run(fmax=0.005)
```

```python
view(water)
```
